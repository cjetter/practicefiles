```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline 
# show the plot inline
#matplotlib.style.use('ggplot') #for a prettier style of visualization
```

```python
aapl=pd.read_csv("AAPL.csv") # read a CSV
aapl.head()
aapl.set_index('Date',drop=True,inplace=True)
aapl=aapl[::-1] # reverse the frame
aapl.Close.plot() # plot one column at a time, much better
```

```python
aapl['Adj Close'].plot()
```

```python
aapl['Adj Close'].plot.bar()
```

```python
df = pd.DataFrame(np.random.randint(100,size=(5,2)), columns=['male','female'])
df.head()
```

```python
df.plot(kind='bar',x=None, y='lable') # vertical bar
```

```python
df.plot.barh() # horizontal bar
```

```python
df.plot.bar(stacked=True) # stacked bar plot
```

```python
df = pd.DataFrame(np.random.randint(100,size=(100,2)), columns=['male','female'])
df['score']=df.male*0.3+df.female*0.7
df.plot.scatter(x='male', y='female',s=df.score)
```
