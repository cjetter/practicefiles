## 2. Select Data

### Convert to Numpy array

```python
data.values
```

### By Column

```python
data['CustomerName'] # column name is case-sensitive
data[['City','CustomerName']] # select by column names, order matters
data[[0,2]] # select by the number of columns, order matters
data.CustomerName # Case-sensitive. Every column is also an attribute of the DataFrame
```

### By Row

```python
data[0:4] # Slicing selection by index, [A, B)

#data[0] # will NOT work
data.ix[0] # will work
data.ix[[0,3]] # multiple selection
data.ix[3:10] # also supports slicing
```

### By Column and Row

```python
print(data.ix[0])
data.ix[0,4] # use column numbers
data.ix[0:3,4:7] # slicing
data.ix[0,'City'] # select by ROW and COL names
data.ix[0:3,'CustomerName':'City'] # you can even slice with column names! Note: last item included
data.ix[:,'CustomerName':'City'].head() # you can select anything with .ix()
```

### By Condition

```python
data.ix[data.CustomerID>20] # select by comparing numbers
data.ix[data.Country=='Germany'] # select by comparing strings
data.ix[data.Country.isin(['Germany','Mexico'])] # more strings
```

### Exercise

Try to use DataFrame.ix() to mimic the following operations:
    1.data[['City','CustomerName']] # select by column names, order matters

    2.data[[0,1]] # select by the number of columns, order matters

    3.data[0:4] # Slicing selection by index, [A, B)

    4.From record No.10-15, select only their name and city.

    5.Select the names of the first and the last customer.

### Solution

```python
# 1.
data.ix[:,['City','CustomerName']] # all the rows, the two columns
# 2.
data.ix[:,[0,1]]
# 3.
data.ix[0:4]
# 4.
data.ix[10:15,['CustomerName','City']]
# 5.
# data.ix[[0,-1],'CustomerName'] # will NOT work
total=len(data) # find the total number of rows
data.ix[[0,total-1],'CustomerName']
# or you can print and count the rows
data.ix[[0,data.tail(1).index.tolist()[0]]] # a clever but hard-to-read answer
```
