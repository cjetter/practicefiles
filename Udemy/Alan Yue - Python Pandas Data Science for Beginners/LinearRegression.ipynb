```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline 
# show the plot inline
matplotlib.style.use('ggplot') #for a prettier style of visualization
```

```python
df=pd.read_excel('commercial rent.xlsx')
df.head()
```

```python
df.describe()
```

```python
# Y = a + bX
import statsmodels.api as sm
```

```python
Y = df.rental_rates
X = df.iloc[:,1:]
X = sm.add_constant(X)
X.head()
```

```python
est = sm.OLS(Y,X)
est = est.fit()
est.summary()
```

```python
# to access a specific attribute, you can pass attribute vs passing the summary table

est.rsquared
```

```python
est.nobs  # number of observations
est.df_model
```
