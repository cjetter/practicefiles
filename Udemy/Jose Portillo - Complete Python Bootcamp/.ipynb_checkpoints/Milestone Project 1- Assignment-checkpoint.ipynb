```python
## Create a Tic Tac Toe game
'''
Requirements:
1) 2 players should be able to play the game (both sitting at the same computer)
2) The board should be printed out every time a player makes a move
3) You should be able to accept input of the player position and then place a symbol on the board

'''
## global variables
spaces = []
placeholder = '___'
for i in range(9):
    spaces.append(placeholder)

player_one_turn = True
winner = False

# create dictionary to map row,col coordinates to index in spaces row
coord_dict = {}
t = 0
for i in range(1,4):
    for n in range(1,4):
        coord_dict[(i,n)]=t
        t+=1
    
## function to introduce and setup game
def game_setup():    
    print ("\nWELCOME TO TIC TAC TOE\n\nPlayer 1 is X's\nPlayer 2 is O's\n")
    print_board()

## function to print current board    
def print_board():
    '''
    function to print out the board (list split into three rows)
    '''
    print('-'*19)
    print('|',spaces[0],'|',spaces[1],'|',spaces[2],'|')
    print('-'*19)
    print('|',spaces[3],'|',spaces[4],'|',spaces[5],'|')
    print('-'*19)
    print('|',spaces[6],'|',spaces[7],'|',spaces[8],'|')
    print('-'*19)
        
def board_entry():
    global spaces, player_one_turn
    if player_one_turn == True: 
        player_no=1
        marker = '_X_'
    else: 
        player_no=2
        marker = '_O_'
        
    print("\n#######  Player %s's Turn #######\n"%player_no)
    print_board()
    
    while True:
        try:
            row = int(input('Player %s: Please select an empty row (1, 2, or 3) where you would like to place your X: '%player_no,))
            if row == 1: row_board = spaces[0:3]
            elif row == 2: row_board = spaces[3:6]
            elif row == 3: row_board = spaces[6:9]
            else: row_board=None
            if placeholder not in row_board: 
                print('Illegal move. Row contains no empty spaces. Please try again.')
                continue
            else: break
        except:
            print('Player %s: Please verify your selection. It must be an integer 1, 2, or 3 and represent a row with an available space: '%player_no)
            continue
    while True:
        try:
            col = int(input('Player %s: Please select an empty column (1, 2, or 3) within row %s where you would like to place your X: '%(player_no,row),))
            if spaces[int(coord_dict[(row,col)])]!=placeholder: 
                print('Illegal move. Location is not empty. Please try again.')
                continue
            else: break
        except:
            print('Player %s: Please verify your selection. It must be an integer 1, 2, or 3 and represent an empty column within row %s: '%(player_no,row))
            continue
    index = int(coord_dict[(row,col)])
    spaces[index]=marker
    
    ## checking rows & columns
    for x in range(3):
        y = x*3
        if spaces[y]==spaces[y+1]==spaces[y+2]==marker*3:
            winner = True
            print_board()
        if spaces[x]==spaces[x+3]==spaces[x+6]==marker*3:
            winner = True
            print_board()

    ## checking diagonals
    if spaces[0]==spaces[4]==spaces[8]==marker*3:
        winner = True
        print_board()
    elif spaces[6]==spaces[4]==spaces[2]==marker*3:
        winner = True
        print_board()
    elif: 
    
    player_one_turn = not player_one_turn

# call functions to run game until winner or error              
game_setup()
while not winner:
    board_entry()

print('\n\n###### PLAYER %s WINS!!! ########\n\n'%player_no)
print('\n\n###### END OF GAME ########\n\n')
```

```python
"""tictactoe game for 2 players
from blogpost: http://thebillington.co.uk/blog/posts/writing-a-tic-tac-toe-game-in-python by  BILLY REBECCHI,
slightly improved by Horst JENS"""
from __future__ import print_function

choices = []

for x in range (0, 9) :
    choices.append(str(x + 1))

playerOneTurn = True
winner = False

def printBoard() :
    print( '\n -----')
    print( '|' + choices[0] + '|' + choices[1] + '|' + choices[2] + '|')
    print( ' -----')
    print( '|' + choices[3] + '|' + choices[4] + '|' + choices[5] + '|')
    print( ' -----')
    print( '|' + choices[6] + '|' + choices[7] + '|' + choices[8] + '|')
    print( ' -----\n')

while not winner :
    printBoard()

    if playerOneTurn :
        print( "Player 1:")
    else :
        print( "Player 2:")

    try:
        choice = int(input(">> "))
    except:
        print("please enter a valid field")
        continue
    if choices[choice - 1] == 'X' or choices [choice-1] == 'O':
        print("illegal move, plase try again")
        continue

    if playerOneTurn :
        choices[choice - 1] = 'X'
    else :
        choices[choice - 1] = 'O'

    playerOneTurn = not playerOneTurn

    for x in range (0, 3) :
        y = x * 3
        if (choices[y] == choices[(y + 1)] and choices[y] == choices[(y + 2)]) :
            winner = True
            printBoard()
        if (choices[x] == choices[(x + 3)] and choices[x] == choices[(x + 6)]) :
            winner = True
            printBoard()

    if((choices[0] == choices[4] and choices[0] == choices[8]) or 
       (choices[2] == choices[4] and choices[4] == choices[6])) :
        winner = True
        printBoard()

print ("Player " + str(int(playerOneTurn + 1)) + " wins!\n")
```
