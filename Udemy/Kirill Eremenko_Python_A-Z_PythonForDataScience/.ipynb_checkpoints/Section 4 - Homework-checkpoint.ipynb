### Function for plot

```python
def plotit(data,playerlist=Players):
    for name in playerlist:
        plt.plot(data[Pdict[name]],c=Cdict[name],ls='--',marker='s',ms=7,label=name)
    plt.xticks(list(range(len(Players))),Seasons)
    plt.legend(loc ='upper left', bbox_to_anchor=(1,1))
    plt.show()    
```

### Data Visualization

#### Observations:
- Dwight Howard historically has drawn more fouls while Joe Johnson has drawn less
- Difference likely relates to the positions played (center vs shooting guard)

---

```python
# Free Throw Accuracy - how good off the FT line?
plotit(FieldGoals/FieldGoalAttempts)  ## on the floor
plotit(FreeThrows/FreeThrowAttempts)  ## on the FT line
```

#### Observations:
- Dwight Howard historically has a very low free throw completion percentage despite drawing a higher number of fouls
- Howard also has a much higher accuracy rating for field goals (likely because he dunks a lot)
- Durant and Paul consistently make foul shots

---

```python
# Player Shooting Style - 2 or 3?
plotit(Points/FieldGoals)
plotit((Points-FreeThrows)/FieldGoals)
```

#### Observations:
- Dwight Howard is an inside player who predominantly makes 2pt baskets
- Joe Johnson is an outside player who shoots more from the perimeter
- Kevin Durant, due to his his Free Throw percentage, intially appears to make a higher degree of 3pts though normalizing for the effect of the FT, we see he shoots closer to 2pt
- Derrick Rose looks to have had rough 2008 and 2009 seasons from the court despite his improved ft percentage
