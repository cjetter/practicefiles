---

```python
## comparison of average life expectancy by region and % change from 1960 to 2013
regions = list(df.Region.unique())

d2 = {x:(np.mean(list(df[df.Region==x]['Expectancy1960'])),\
         np.std(list(df[df.Region==x]['Expectancy1960'])),\
         np.mean(list(df[df.Region==x]['Expectancy2013'])),\
         np.std(list(df[df.Region==x]['Expectancy2013']))) for x in regions}
df3 = pd.DataFrame(data=d2).T
df3.columns =['AvgExpectancy_1960','Std_1960','AvgExpectancy_2013','Std_2013']
df3['Change_Avg_%'] = (df3.AvgExpectancy_2013-df3.AvgExpectancy_1960)/df3.AvgExpectancy_1960 *100
df3['Change_Std'] = df3.Std_2013-df3.Std_1960
mean_1960 = np.mean(df3.AvgExpectancy_1960)
print(mean_1960)
df3['Relative_to_Group_Mean_1960'] = df3.AvgExpectancy_1960-mean_1960
mean_2013 = np.mean(df3.AvgExpectancy_2013)
print(mean_2013)
df3['Relative_to_Group_Mean_2013'] = df3.AvgExpectancy_2013-mean_2013
df3
```

#### Observations
- All regions saw an increase in the average life expectancy
- There looks to be a correlation between a higher life expectancy and lower birth rate (no need to produce more to ensure survival of family)
- The African region and European regions represent the extremities
- All regions but Africa have narrowed the regional disparity among expectancies
