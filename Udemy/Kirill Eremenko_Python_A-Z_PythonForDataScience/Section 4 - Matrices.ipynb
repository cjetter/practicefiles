### Section 4 - Matrices

#### Creating a matrix
- You can use the np.reshape('array','new_shape','C'/'F')
- The 'C' stands for C programming language and first populates the rows (left, right, down) while the 'F' stands for Fortran and first populates the columns (top, down, over)
- You can use np.array and pass in rows of data
<img src='Section 4 - image.png' style='width: 700px;' align='center'/>

```python
import numpy as np
mydata = np.arange(0,20)   ## could also pass in a list of numbers 'list(range(20))'
print(mydata)
```

```python
## Matrix created according to C style - rows filled first
MATR1 = np.reshape(mydata,(5,4))
MATR1
```

```python
# Exercise - access the number 10 in the matrix
var = MATR1[2][2]     # or MATR1[2,2]
var
```

```python
## Matrix created according to Fortran style - columns filled first
MATR2 = np.reshape(mydata,(5,4),'F')
MATR2
```

```python
# Exercise - access the number 10 in the matrix2
var2 = MATR2[0][2]     # or MATR2[0,2]
var2
```

```python
## OOP Concept:
# MATR1 = np.reshape(mydata,(5,4))
# Given 'mydata' is already a NumPy array object, you don't have to go get the package and pass it in to the
# reshape function. Instead you can call the reshape function from within the object itself.

MATR3 = mydata.reshape((10,2))
MATR3

## NOTE: would not work if you originally created mydata as a list to pass into the np.reshape function
```
