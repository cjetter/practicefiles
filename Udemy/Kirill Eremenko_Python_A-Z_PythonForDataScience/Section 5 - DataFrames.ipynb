### Basic DataFrame Operations

```python
## Mathematical operations
results = stats.BirthRate * stats.InternetUsers
results.head()
```

```python
## add a column
stats['myCalc'] = stats.BirthRate * stats.InternetUsers          ## similar to dictionary additions
stats.head()
```

```python
## Unlike R, which is a vectoring language such that it knows to recycle a vector (ie.list) to populate all rows 
# by copy-pasting the vector until it has enough values to populate all rows, you must pass in a series of values 
# equal to the number of rows

#stats['NewCol'] = [1,2,3,4,5]           # generates error in python while in R would work
stats['NewCol'] = [(1,2)]*len(stats)     # code to add new row based on outside values (ie. not derived from existing df columns)
stats
stats['NewCol'] = [1]*len(stats)         # will overwrite existing row values
stats.head()
```

```python
## removing a column
stats.drop(['myCalc'],1)      # pass in a 1 to specify columns (otherwise defaults to 0 for rows)
                              # note - will only return a copy of the editted df. must copy df on itself or specify
                              # 'inplace' as True
print(stats.head())
stats = stats.drop(['myCalc'],1)
print(stats.head())
stats.drop(['NewCol'],1,inplace=True)
stats.head()
```

### Filtering DataFrames

```python
# filter is  more about isolating rows (similar to dropdown in excel)
stats.InternetUsers < 2               # in which rows of the Interent Users col of DF is condition met
Filter = stats.InternetUsers < 2      # pass filtered object into variable
stats[Filter]                         # only returns rows of DF where True
stats[stats.InternetUsers < 2]        # combine condition and filter into one step
```

```python
# sorting by ==
stats[stats.IncomeGroup == 'High income']
```

```python
# filter on multiple criteria
filter1 = stats.InternetUsers<2
filter2 = stats.BirthRate>40
stats[filter1 & filter2]
#stats[(stats.InternetUsers<2) and (stats.BirthRate>40)]  # will not work as python can only compare one value 
                                                          # (vs list of Falses and Trues)
stats[(stats.InternetUsers<2) & (stats.BirthRate>40)]     # ampersand will compare bitwise
                                                          # must include the parentheses to avoid error
```

```python
# how to get the unique categories
list(stats.IncomeGroup.unique())         # similar to 'remove duplicate values' in Excel
```

```python
# find details on Malta
details = stats[stats.CountryName == 'Malta']
details
```

### use of at() and .iat() to accesss individual element

```python
# '.at[]' used for labels = important: even integers are treated as labels
# '.iat[]' used for integer locations [row, col]
```

```python
# search by index
print(stats.iat[19,0])                     
stats.iat[0,0] == stats.iloc[0][0]       # can also use .iloc
```

```python
# search by label
print(stats.at[2,'BirthRate'])
print(stats.at[19,'CountryName'])
```

```python
## why we need .at
sub10 = stats[::10]
print(sub10)
print(sub10.iat[10,0])              # with .iat the 10 is treated as an index position
print(sub10.at[10,'CountryName'])   # with .at the 10 is treated as a string label equal to '10' tho input as int
# print(sub10.at[11,0]) # will generate error as label 11 is not included in sub10 df
print(sub10.iat[11,0])  # will print as 11 as an index is included in sub10 df
```

### Keyword Arguments in Python (using seaborn example)

- given that seaborn is built on top of matplotlib, you can access underlying kwargs availble for plt.scatter or plt.line by accessing a dictionary through the seaborn kwargs of 'scatter_kws=None' or 'line_kws=None'
- seaborn function acts as a wrapper of matplotlib function ('sns.lmplot' is wrapper of 'plt.scatter')
    - upon receiving a dictionary, seaborn will unwrap it and pass it to matplotlib for interpretation


```python
# linear model plot of internet users to birthrate
# increasing marker size

vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',\
                  fit_reg=False,hue='IncomeGroup',size=8,aspect=2,\
                  scatter_kws={'s':150})  # by passing in a value via dictionary for the underlying kwarg 's' of 
                                          # matplotlib, you are able to access optionality not immediately
                                          # available with Seaborn
```
